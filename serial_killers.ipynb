## Observations 
To address the hypothysis we need number of killings per year. We dont have that data, all we have is killings across a 
time period so we are going to: 
    1. Calculate a duration using the from and to date.
    1. Use duration and victim count to calculate an estimate of killings per year. 
    
    
We should draw a chart to show the trends from this data. 

Lets take a look at the data and see if there is anything we can clean up. 

```python
skm
```

There are some dangling single quotes and lower case errors in the names that need to be cleaned from the table. 

```python
skm['Name']= skm['Name'].str.strip("'").str.title()
skm
```

Removed any killers listed with < 1 victim count.

```python
skm = skm[skm['Victim Count'] > 0]
skm
```

turned off default warning in pandas using code from stack overflow.

```python
# skm['duration'] = 1 + skm['To'].astype(int) - skm['From'].astype(int)
# skm
pd.options.mode.chained_assignment = None  # default='warn'
skm['duration'] = skm ['From'].astype(int) - skm['To'].astype(int) 
skm
```

```python
skm['rate'] = skm['Victim Count'] /  skm['duration']
skm
```

```python
gender_totals = skm.groupby('Gender')['Victim Count'].sum()
gender_totals
```

```python
gender_totals.plot(kind = 'bar')
```

```python
modus_totals = skm.groupby('Category')['Victim Count'].sum()
modus_totals
```

```python
modus_totals.plot(kind = 'pie')
```
